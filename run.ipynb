Our model expects a list of 30 numerical features as input — one row from our dataset (excluding the label Class). We can get this from our local creditcard.csv.

```python
import pandas as pd
import requests
import json
```

```python
# Let's have look at the data
df = pd.read_csv("data/creditcard.csv")
df.head()
```

```python
# To test the real-time inference endpoint, we'll use one of the transactions in the dataset. We can change the numbers thogh.
sample_features = df.drop(columns=['Class']).iloc[0].tolist()
print(sample_features)
```

```python
# Modify the url if according to yours
url = "https://i9j9jpazs3.execute-api.us-east-2.amazonaws.com/predict"

payload = {
    "features": sample_features  # ← insert real values
}

response = requests.post(url, json=payload)
print(response.json())
```
